# 01 Data Setup

Use this notebook to verify paths, inspect dataset splits, and preview a few real images before training anything.

```python
from pathlib import Path
import os
import sys

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
scripts_dir = ROOT / "scripts"
if str(scripts_dir) not in sys.path:
    sys.path.insert(0, str(scripts_dir))
os.chdir(ROOT)
ROOT
```

```python
from collections import defaultdict
from pprint import pprint

data_root = ROOT / "data_final"
summary = defaultdict(dict)

for split in ["train", "val", "test"]:
    split_dir = data_root / split
    for class_dir in sorted([p for p in split_dir.iterdir() if p.is_dir()]):
        summary[split][class_dir.name] = sum(1 for p in class_dir.iterdir() if p.is_file())

pprint(dict(summary))
```

## Real image preview

The cell below shows the first image found in each training class.

```python
from IPython.display import display
from PIL import Image

for class_name in ["apple", "banana", "orange"]:
    sample_path = sorted((data_root / "train" / class_name).glob("*"))[0]
    print(class_name, sample_path.name)
    display(Image.open(sample_path))
```
